```python
# 7 -> Write a program that handles exceptions in a recursive factorial function.
def func(n):
  try:
    if n<0:
      raise ValueError("give a positive number")
    if n==0 or n==1:
      return 1
    else:
      return n * func(n-1)
  except RecursionError as e:
    print("error",e)
  except ValueError as e:
    print("error",e)
func(5)
```

```python
# 8. Write a program that simulates transaction rollback in a banking system using exceptions.
balance = 1000
print("Initial Balance:", balance)

try:
    backup = balance

    # transaction
    balance -= 500
    balance += -100

    if balance < 0 or -100 <= 0:
        raise Exception("Transaction Error!")

    print("Transaction Successful!")

except Exception as e:
    print("Transaction Failed:", e)
    balance = backup

print("Final Balance:", balance)
```

```python
# 9. Write a program to validate credit card numbers and raise exceptions for invalid inputs.
def validate_credit_card(number):
    try:
        number = number.replace(" ", "")
        if not number.isdigit():
            raise ValueError("Credit card number must contain only digits.")
        if len(number) < 13 or len(number) > 16:
            raise ValueError("Credit card number must be 13 to 16 digits long.")

        print("Credit card number is valid.")
    except ValueError as e:
        print("Invalid credit card number:", e)
validate_credit_card("1234 5678 9012 3456")
```
