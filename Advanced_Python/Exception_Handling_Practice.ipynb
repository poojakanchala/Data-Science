```python
# 7 -> Write a program that handles exceptions in a recursive factorial function.
def func(n):
  try:
    if n<0:
      raise ValueError("give a positive number")
    if n==0 or n==1:
      return 1
    else:
      return n * func(n-1)
  except RecursionError as e:
    print("error",e)
  except ValueError as e:
    print("error",e)
func(5)
```
